```python
import pandas as pd
```

```python
dfwine = pd.read_csv('/content/winemag-data-130k-v2.csv')
dfwine.dropna(inplace = True)
dfwine.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
print(dfwine.dtypes)
numerical_df = dfwine.select_dtypes(include='number')

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()
correlation_matrix
```

```python
dfwine[['price','points']].corr()
```

```python
# prompt: build a bar chart with the x and y data points

# Extract the 'price' and 'points' columns as lists
x_data = dfwine['price'].tolist()
y_data = dfwine['points'].tolist()

# Create the bar chart
plt.bar(x_data, y_data)

# Set the labels for the x and y axes
plt.xlabel('Price')
plt.ylabel('Points')

# Set the title of the chart
plt.title('Wine Price vs. Points')

# Display the chart
plt.show()
```
